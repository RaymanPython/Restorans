Константные переменные

```python
INDEX_COL = 0
AVERAGE_BILL = "average_bill"
AVERAGE_BILL_FILLNA =  "average_bill_filna"
RUBRIC = "rubric"
```

```python
data = pd.read_csv("datasets_eat_places.csv", index_col=INDEX_COL)
```

```python
data.head()

```

```python
counts = 0
for i, j in data.iterrows():
    if j[AVERAGE_BILL] > 2500 or j[AVERAGE_BILL] == np.nan:
        counts += 1
print(counts)
```

```python
data[AVERAGE_BILL]
```

```python
data[AVERAGE_BILL_FILLNA] = data[AVERAGE_BILL]
data[AVERAGE_BILL_FILLNA].fillna(0, inplace=True)
avarage_price = sum(data[AVERAGE_BILL_FILLNA] / len(data[AVERAGE_BILL_FILLNA])) #средняя цена
```

```python
avarage_price
```

```python
data[AVERAGE_BILL].fillna(avarage_price, inplace=True)
```

```python
data
```

Легко заметить посмотрев на предыдущие таблицы, что NaN мы убрали из среднего чека

```python
data[AVERAGE_BILL] = data[AVERAGE_BILL].map(lambda x: avarage_price if (x > 2500) else x)
```
